## Future: Wallet Credit Scoring Model (Dataset can be added later)

This section adds a **future-ready** training pipeline for wallet credit scoring.

### Planned training features
- Wallet balance (ETH + USDT value in USD)
- Transaction history (transaction count)
- NFT holdings (NFT ownership volume)
- Account age (estimated wallet age)
- Network diversity (token diversity)

### Expected dataset schema (CSV)
- `wallet_balance_usd`
- `transaction_count`
- `nft_ownership_volume`
- `account_age_days`
- `token_diversity`
- `credit_score` (target label)

If a real dataset is not present yet, the code below builds a small synthetic dataset so the notebook remains runnable.

```python
from pathlib import Path
import csv
import random
from statistics import mean

FEATURE_COLUMNS = [
    "wallet_balance_usd",
    "transaction_count",
    "nft_ownership_volume",
    "account_age_days",
    "token_diversity",
]
TARGET_COLUMN = "credit_score"
DATASET_PATH = Path("credit_scoring_dataset.csv")  # Add real dataset later

# Wallet balance is intentionally the highest-weight feature.
PRIOR_FEATURE_WEIGHTS = {
    "wallet_balance_usd": 0.40,
    "transaction_count": 0.20,
    "nft_ownership_volume": 0.15,
    "account_age_days": 0.15,
    "token_diversity": 0.10,
}


def _synthetic_row() -> dict:
    wallet_balance_usd = random.uniform(50, 150_000)
    transaction_count = random.randint(1, 5000)
    nft_ownership_volume = random.randint(0, 200)
    account_age_days = random.randint(30, 3650)
    token_diversity = random.randint(1, 30)

    # Synthetic label formula (temporary placeholder)
    score = (
        300
        + min(wallet_balance_usd / 1000, 220)
        + min(transaction_count / 30, 170)
        + min(nft_ownership_volume * 1.2, 90)
        + min(account_age_days / 25, 110)
        + min(token_diversity * 6, 80)
    )
    score = int(max(300, min(850, round(score))))

    return {
        "wallet_balance_usd": round(wallet_balance_usd, 2),
        "transaction_count": transaction_count,
        "nft_ownership_volume": nft_ownership_volume,
        "account_age_days": account_age_days,
        "token_diversity": token_diversity,
        "credit_score": score,
    }


def build_or_load_dataset(path: Path, synthetic_rows: int = 250) -> list[dict]:
    if path.exists():
        rows: list[dict] = []
        with path.open("r", newline="", encoding="utf-8") as f:
            reader = csv.DictReader(f)
            for row in reader:
                rows.append(
                    {
                        "wallet_balance_usd": float(row["wallet_balance_usd"]),
                        "transaction_count": float(row["transaction_count"]),
                        "nft_ownership_volume": float(row["nft_ownership_volume"]),
                        "account_age_days": float(row["account_age_days"]),
                        "token_diversity": float(row["token_diversity"]),
                        "credit_score": float(row["credit_score"]),
                    }
                )
        return rows

    synthetic = [_synthetic_row() for _ in range(synthetic_rows)]
    with path.open("w", newline="", encoding="utf-8") as f:
        writer = csv.DictWriter(f, fieldnames=[*FEATURE_COLUMNS, TARGET_COLUMN])
        writer.writeheader()
        writer.writerows(synthetic)
    return synthetic


def _feature_min_max(rows: list[dict]) -> dict[str, tuple[float, float]]:
    stats: dict[str, tuple[float, float]] = {}
    for col in FEATURE_COLUMNS:
        values = [float(r[col]) for r in rows]
        stats[col] = (min(values), max(values))
    return stats


def _normalize_feature(value: float, min_val: float, max_val: float) -> float:
    spread = max_val - min_val
    if spread <= 1e-9:
        return 0.0
    norm = (value - min_val) / spread
    return max(0.0, min(1.0, norm))


def _weighted_index(row: dict, weights: dict, min_max_stats: dict[str, tuple[float, float]]) -> float:
    return sum(
        weights[col] * _normalize_feature(float(row[col]), min_max_stats[col][0], min_max_stats[col][1])
        for col in FEATURE_COLUMNS
    )


def train_weighted_credit_model(rows: list[dict], prior_weights: dict) -> tuple[dict, dict, float, float]:
    # Keep the feature importance policy fixed (wallet highest),
    # and learn score calibration from data: score ≈ intercept + scale * weighted_index.
    min_max_stats = _feature_min_max(rows)

    x = [_weighted_index(r, prior_weights, min_max_stats) for r in rows]
    y = [float(r[TARGET_COLUMN]) for r in rows]

    mean_x = mean(x)
    mean_y = mean(y)
    var_x = sum((v - mean_x) ** 2 for v in x)
    cov_xy = sum((a - mean_x) * (b - mean_y) for a, b in zip(x, y))

    scale = cov_xy / var_x if var_x > 1e-12 else 0.0
    intercept = mean_y - (scale * mean_x)

    return prior_weights, min_max_stats, scale, intercept


def predict_credit_score(
    wallet_features: dict,
    weights: dict,
    min_max_stats: dict[str, tuple[float, float]],
    scale: float,
    intercept: float,
) -> int:
    idx = _weighted_index(wallet_features, weights, min_max_stats)
    raw = intercept + (scale * idx)
    return int(max(300, min(850, round(raw))))


rows = build_or_load_dataset(DATASET_PATH)
weights, min_max_stats, scale, intercept = train_weighted_credit_model(rows, PRIOR_FEATURE_WEIGHTS)

sample_wallet = {
    "wallet_balance_usd": 12500,
    "transaction_count": 420,
    "nft_ownership_volume": 14,
    "account_age_days": 920,
    "token_diversity": 9,
}
predicted_score = predict_credit_score(sample_wallet, weights, min_max_stats, scale, intercept)

print(f"Dataset path: {DATASET_PATH.resolve()}")
print(f"Rows used for training: {len(rows)}")
print("Model status: Wallet-balance-priority weighted model calibrated on dataset.")
print("Sample wallet predicted score:", predicted_score)
print("Configured feature weights (wallet is highest):")
for feature, weight in sorted(weights.items(), key=lambda item: item[1], reverse=True):
    print(f"  - {feature}: {weight:.2f}")
print(f"Calibration scale: {scale:.4f}")
print(f"Calibration intercept: {intercept:.4f}")
```
